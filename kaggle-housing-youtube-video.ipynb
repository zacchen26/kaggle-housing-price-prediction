# feature engineering

House Age (houseage): This feature represents the age of the house at the time of sale, calculated as the difference between the year the house was sold (YrSold) and the year the house was built (YearBuilt). This is useful to understand how old the house is, as older houses might need more maintenance or could have historical value that affects their price.

```python
train_df['houseage'] = train_df['YrSold'] - train_df['YearBuilt']
test_df['houseage'] = test_df['YrSold'] - test_df['YearBuilt']
```

House Remodel Age (houseremodelage): Similar to houseage, this feature calculates the age of the last remodel at the time of sale. It's the difference between the year the house was sold (YrSold) and the year the last remodeling was done (YearRemodAdd). It helps capture the recency of updates which can significantly affect the house's market value.

```python
train_df['houseremodelage'] = train_df['YrSold'] - train_df['YearRemodAdd']
test_df['houseremodelage'] = test_df['YrSold'] - test_df['YearRemodAdd']
```

Total Square Feet (totalsf): This is the total finished square footage of the house including all floors and the basement area. It sums up the square footage of the first and second floors (1stFlrSF and 2ndFlrSF), and both finished areas of the basement (BsmtFinSF1 and BsmtFinSF2). Total square footage is a primary factor in determining a property's value.

```python
train_df['totalsf'] = train_df['1stFlrSF'] + train_df['2ndFlrSF'] + train_df['BsmtFinSF1'] + train_df['BsmtFinSF2']
test_df['totalsf'] = test_df['1stFlrSF'] + test_df['2ndFlrSF'] + test_df['BsmtFinSF1'] + test_df['BsmtFinSF2']
```

Total Living Area (totalarea): This feature sums the above-grade (ground) living area (GrLivArea) and the total basement area (TotalBsmtSF). This combined feature provides a comprehensive measure of the total usable living space, which is a critical variable in real estate pricing.


```python
train_df['totalarea'] = train_df['GrLivArea'] + train_df['TotalBsmtSF']
test_df['totalarea'] = test_df['GrLivArea'] + test_df['TotalBsmtSF']
```

Total Bathrooms (totalbaths): This feature calculates the total number of bathrooms, including full and half baths, in the house. Full bathrooms (BsmtFullBath, FullBath) count as one each, and half baths (BsmtHalfBath, HalfBath) count as 0.5. The number of bathrooms is a key feature in home valuation.


```python
train_df['totalbaths'] = train_df['BsmtFullBath'] + train_df['FullBath'] + 0.5 * (train_df['BsmtHalfBath'] + train_df['HalfBath']) 
test_df['totalbaths'] = test_df['BsmtFullBath'] + test_df['FullBath'] + 0.5 * (test_df['BsmtHalfBath'] + test_df['HalfBath']) 
```

Total Porch Area (totalporchsf): This combines all different types of porch areas including open porch, enclosed porch, three-season porch, screen porch, and wood deck area (OpenPorchSF, 3SsnPorch, EnclosedPorch, ScreenPorch, WoodDeckSF). The total porch area adds to the aesthetic and functional appeal of the house, potentially affecting its desirability and value.


```python
train_df['totalporchsf'] = train_df['OpenPorchSF'] + train_df['3SsnPorch'] + train_df['EnclosedPorch'] + train_df['ScreenPorch'] + train_df['WoodDeckSF']
test_df['totalporchsf'] = test_df['OpenPorchSF'] + test_df['3SsnPorch'] + test_df['EnclosedPorch'] + test_df['ScreenPorch'] + test_df['WoodDeckSF']
```

```python
num_cols = train_df.select_dtypes(include=['int64', 'float64']).columns
num_cols = num_cols.drop('SalePrice')
```

```python
num_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='mean')), # to impute the missing value with mean
    ('scaler', StandardScaler()) # value/mean
])
```

```python
# Ordinal Encoding considers the ranking. like high school0; bachelor1; master2;phd3
ode_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('ode', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])
```

```python
# One-Hot Encoding ignores ranking. Red: [1, 0, 0], Blue: [0, 1, 0], Green: [0, 0, 1]
ohe_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])
```

```python
col_trans = ColumnTransformer(transformers=[
    ('num_p', num_pipeline, num_cols),
    ('ode_p', ode_pipeline, ode_cols),
    ('ohe_p', ohe_pipeline, ohe_cols),
    ],
    remainder='passthrough', 
    n_jobs=-1)
```

```python
pipeline = Pipeline(steps=[
    ('preprocessing', col_trans)
])
```
